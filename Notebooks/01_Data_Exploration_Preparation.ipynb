```python
mfde["Pounds"] = pd.to_numeric(
    mfde["Pounds"],
    errors="coerce"
)

mfde["Value"] = pd.to_numeric(
    mfde["Value"],
    errors="coerce"
)
```

```python
species_summary = (
    mfde.groupby("Species Name")
    .agg(
        first_year=("Year", "min"),
        last_year=("Year", "max"),
        years=("Year", "nunique"),
        total_pounds=("Pounds", "sum"),
        confidential_years=("Pounds", lambda x: x.isna().sum())
    )
)

species_summary["pct_confidential"] = (
    species_summary["confidential_years"] /
    species_summary["years"] * 100
)

species_summary = species_summary.sort_values(
    "total_pounds",
    ascending=False
)

species_summary.head(25)
```
